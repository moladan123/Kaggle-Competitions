```python
import pandas as pd
import numpy as np

ROOT = "/Volumes/Unititled/Jan-Apr-2019/Random/KaggleCompetitions/titanicData/titanic/"

test = pd.read_csv(ROOT + "test.csv")
train = pd.read_csv(ROOT + "train.csv")

train.head()
```

```python
# Preprocessing, We encode everything into a numeric matrix

def preprocess(data):
    # drop a bunch of the information
    # some of this stuff could potentially be usedful, but for the time being will be ignore
    data = data.drop(["PassengerId", "Name", "Ticket", "Cabin", "Embarked"], axis=1)

    # Make sex a binary number
    data["Sex"] = [1 if person == 'male' else 0 for person in data["Sex"]]

    # assumes average age if not given
    data = data.fillna(value=25)
    
    y = None
    try:
        # get the output vector
        y = data["Survived"]
        data = data.drop("Survived", axis=1)
        data = data.values
    except:
        pass

    return (data, y)


data, y = preprocess(train)
data
```

## Training

Now that our data is in a more convenient format, we can now build the classifier.

```python
from sklearn import svm

# creating a support vector machine
clf = svm.SVC(gamma=0.001, C=100.)

clf.fit(data, np.array(y))
```

```python
# Predict the value here
data, _ = preprocess(test)

predictions = list(clf.predict(data))
labels = list(test["PassengerId"])

out = pd.DataFrame({"PassengerId": labels, "Survived": predictions})

out.to_csv(ROOT + "out.csv")

out.head(n=10)
```
